```python
from pytrends.request import TrendReq
pytrends=TrendReq(hl="en-US",tz=330)
```

```python
kw_list = ['blazers','Striped shirt','OverSized','Designer Saree'] # list of keywords to get data 

pytrends.build_payload(kw_list, cat=0, timeframe='now 7-d') 
```

```python
#1 Interest over Time
data = pytrends.interest_over_time() 
data = data.reset_index() 


import plotly.express as px

fig = px.line(data, x="date", y=['blazers','Striped shirt','OverSized','Designer Saree'], title='Keyword Web Search Interest Over Time')
fig.show() 
```

```python
d1=data.mean()
```
