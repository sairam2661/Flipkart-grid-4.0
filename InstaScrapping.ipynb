```python
def scrapeprofiles(keyword):

    searchbox = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='Search']")))
    searchbox.clear()
    searchbox.send_keys(keyword)
    time.sleep(5)
    clicker = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//div[contains(text(),"'+keyword+'")]'))).click()
    time.sleep(5)
    
    driver.execute_script("window.scrollTo(0, 4000);")

    images = driver.find_elements_by_tag_name('img')
    images = [image.get_attribute('src') for image in images]
    images = images[2:-2]
    
    fashion = pd.DataFrame(images[:5],columns=['url'])
    
    path = os.getcwd()
    
    counter = 0
    image_name = []
    for i in range(image_count):
        save_as = os.path.join(path+'/current-fashion', keyword[0:] + str(counter) + '.jpg')
        image_name.append(keyword[1:] + str(counter) + '.jpg')
        wget.download(images[i], save_as)
        counter += 1
    fashion['Images'] = image_name
    
    links = []
    links = driver.find_elements_by_tag_name('a')

    def condition(link):
        return '.com/p/' in link.get_attribute('href')

    valid_links = list(filter(condition,links))
    
    links=[]
    for i in range(5):
        link = valid_links[i].get_attribute('href')
        if link not in links:
            links.append(link)
            
    likes = []
    for link in links:
        driver.get(link)
        time.sleep(5)
        try:
            like = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/section/main/div[1]/div[1]/article/div/div[2]/div/div[2]/section[2]/div/div/div/a/div/span"))).text
        except:
            pass
        likes.append(like)
        time.sleep(5)
        
    fashion['Likes'] = likes 
    return fashion
```
